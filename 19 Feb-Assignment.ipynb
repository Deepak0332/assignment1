## Q1. What is Flask Framework? What are the advantages of Flask Framework?

## Answer:
Flask is a micro web framework written in Python. It is classified as a micro framework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where the pre-existing third-party libraries provide common functions. Flask is used to build small to medium-sized web applications, APIs, and complex web apps.

**Advantages of Flask Framework:**

1. Lightweight: Flask is a lightweight framework and is easy to install.

2. Flexibility: Flask allows developers to use any database, template engine, or other libraries as per their choice.

3. Minimalistic: With Flask, you only get what you need. There is no overhead or unnecessary components.

4. Easy to learn: Flask is beginner-friendly and easy to understand. You can quickly build and deploy small web applications with Flask.

5. Extensible: Flask is highly extensible and can integrate with multiple Python libraries.

6. Scalable: Flask can easily handle a large number of users and database records. It can scale up as per the application's needs.

7. Flask provides a ready-to-use development server and flexible integrated

## Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

## Answer:

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

## Q3. What is App routing in Flask? Why do we use app routes?





## Answer:
App routing in Flask is the process of mapping URLs to specific functions or views in a Flask application. It allows the application to handle different requests from clients and return appropriate responses.

We use app routes in Flask to define the behavior of our application when a user visits a specific URL. By defining routes, we can create different pages or endpoints within our application that serve different purposes. For example, we can define a route for the homepage, a route for user login, and a route for displaying user data.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation<br>
Location: India<br>
Contact Detail: 999-999-9999

## Answer:

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation<h1>"

@app.route("/")
def company_details():
    return "<h1>Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999</h1>"

if __name__ == '__main__':
    app.run(port = 8080)
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

## Answer:
In Flask, the url_for() function is used for URL building. It generates a URL for a specific endpoint based on the endpoint name and any parameters passed to it.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def home():
    company_name = 'ABC Corporation'
    location = 'India'
    contact_detail = '999-999-9999'
    welcome_url = url_for('welcome') 
    return f'Company Name: {company_name}<br>Location: {location}<br>Contact Detail: {contact_detail}<br><a href="{welcome_url}">Click here to go to the welcome page</a>'

if __name__ == '__main__':
    app.run(port = 7000)
```
